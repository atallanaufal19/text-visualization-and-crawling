```python
import pandas as pd

# Specify the path to your CSV file
file_path = f"tweets-data/{filename}"

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, delimiter=";")

# Display the DataFrame
display(df)
```

```python
df_text = df[['full_text']]
```

```python
import nltk
from nltk.corpus import stopwords
import re  # Import the re module for regular expressions

# Download NLTK stopwords for Indonesian
nltk.download('stopwords')
stop_words_indonesian = set(stopwords.words('indonesian'))

# Kata-kata tambahan yang ingin dihapus
additional_stopwords = {'ya', 'ini', 'sih', 'nya', 'aja', 'utk', 'jd', 'krn',
                        'tdk', 'sy', 'dgn', 'karena', 'sudah', 'sdh',
                        'td', 'yg', 'ga', 'si', 'gak', 'tp'}

# Gabungkan stopwords dari NLTK dengan kata-kata tambahan yang ingin dihapus
stop_words_indonesian = stop_words_indonesian.union(additional_stopwords)

# Custom function to remove Twitter handles, links, and specified stopwords
def clean_text_custom(text):
    # Remove words containing '@' (Twitter handles)
    text = ' '.join([word for word in text.split() if '@' not in word])
    # Remove words containing 'https' (links)
    text = re.sub(r'https\S+', '', text)
    # Remove specified stopwords
    text = ' '.join([word for word in text.split() if word.lower() not in stop_words_indonesian])
    return text

# Apply the custom function to remove Twitter handles, links, and specified stopwords
df_text['cleaned_text_custom'] = df_text['full_text'].apply(clean_text_custom)
df_text
```

```python
from google.colab import files

# Simpan DataFrame ke dalam file Excel
excel_file_path = 'tweets_data_cleaned.xlsx'
df_text.to_excel(excel_file_path, index=False)

# Unduh file Excel
files.download(excel_file_path)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Concatenate all the cleaned tweets into a single string
all_cleaned_text = ' '.join(df_text['cleaned_text_custom'])

# Generate the word cloud with a maximum of 50 words
wordcloud = WordCloud(width=800, height=400, background_color='white', max_words=50).generate(all_cleaned_text)

# Display the word cloud using matplotlib
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

```
